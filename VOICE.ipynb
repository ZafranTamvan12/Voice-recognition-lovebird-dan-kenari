```python
import librosa
import librosa.core.audio
import librosa.display
import IPython.display as ipd
import os
import pandas as pd
import numpy as np
```

```python
audio_dataset_path = '/content/dataset/'
metadata = pd.read_csv('/content/dataset.csv')
metadata.head()
```

```python
file_name2 =metadata['slice_file_name']
print(file_name2)
```

```python
classification = metadata['class']
print(classification)
```

```python
def features_extractor(file_name):
  audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
  mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=60)
  mfccs_scaled_feature = np.mean(mfccs_features.T, axis=0)

  return mfccs_scaled_feature
```

```python
features = []
for index, row in metadata.iterrows():
    file_name = os.path.join(audio_dataset_path, row['slice_file_name'])
    class_label = row['class']
    data = features_extractor(file_name)
    features.append([data, class_label])
```



1. **Membentuk Array Fitur (X)**:
   - Dalam kode ini, `features` adalah daftar yang berisi pasangan fitur dan kelas.
   - List comprehension `[item[0] for item in features]` digunakan untuk mengambil elemen pertama dari setiap pasangan fitur dan kelas dalam `features`.
   - Setiap elemen pertama dari setiap pasangan fitur dan kelas adalah fitur itu sendiri.
   - Kemudian, `np.array()` digunakan untuk mengonversi daftar fitur menjadi array numpy.
   - Jadi, `X` adalah array numpy yang berisi fitur-fitur dari data.

2. **Membentuk Array Kelas (y)**:
   - Dalam kode ini, lagi-lagi menggunakan list comprehension `[item[1] for item in features]` untuk mengambil elemen kedua dari setiap pasangan fitur dan kelas dalam `features`.
   - Setiap elemen kedua dari setiap pasangan fitur dan kelas adalah kelas atau label.
   - Kemudian, `np.array()` digunakan untuk mengonversi daftar label menjadi array numpy.
   - Jadi, `y` adalah array numpy yang berisi kelas atau label dari data.



```python
X = np.array([item[0] for item in features])
y = np.array([item[1] for item in features])
```

```python
X
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = to_categorical(labelencoder.fit_transform(y))
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
model =Sequential()
model.add(Dense(len(X_train[0]),
                activation='relu', input_shape=X_train[0].shape))
model.add(Dense(len(X_train[0, :]), activation='relu'))
model.add(Dense(len(X_train[0, :]), activation='tanh'))
model.add(Dense(len(y_train[0, :]), activation='sigmoid'))
```

```python
# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Tampilkan ringkasan model
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime

num_epochs = 15
num_batch = 32

checkpointer = ModelCheckpoint(filepath='saved_models/voicerecog.h5',
                               verbose=1, save_best_only=True)
start = datetime.now()
history = model.fit(X_train, y_train, validation_data=(
    X_test, y_test), callbacks=[checkpointer], epochs=num_epochs, batch_size=num_batch)

duration = datetime.now() - start
print('Training completed in time: ', duration)

# Print accuracy
train_accuracy = history.history['accuracy'][-1] * 100
test_accuracy = history.history['val_accuracy'][-1] * 100
print('Train Accuracy: {:.2f}%'.format(train_accuracy))
print('Test Accuracy: {:.2f}%'.format(test_accuracy))
```

```python
prediction_label = model.predict(X_test)
print(prediction_label)
```
